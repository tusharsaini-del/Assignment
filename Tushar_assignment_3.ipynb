```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'E:\6th sem notes\data anl. training\train.csv')
```

```python
df.info()
```

1.Data cleaning and preprocessing

```python
print("First 5 rows of dataset:")
print(df.head())

# Check missing values
print("\nMissing Values in Dataset:")
print(df.isnull().sum())

# Fill missing Age values with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked values with most common value (mode)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop unnecessary columns
df = df.drop(columns=["Name", "Ticket", "Cabin"])

print("\nDataset after cleaning:")
print(df.isnull().sum())
```

2.Univariative analysis

Univariate Analysis: basic theory and example - ToolsheroUnivariate analysis is the most basic form of statistical data analysis, focusing on only one variable at a time to describe, summarize, and identify patterns. It helps detect central tendency (mean, median, mode), dispersion (range, variance, standard deviation), and outliers, often using visual tools like histograms or box plots. 

```python
#Age distribution
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Distribution of Age (Histogram + KDE)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#  Count of Passengers by Sex (Bar Chart)
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df)
plt.title("Count of Passengers by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of Fare using Boxplot

plt.figure(figsize=(8,4))
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution (Boxplot)")
plt.xlabel("Fare")
plt.show()
```

 3. Bivariate Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Pclass vs Fare (Boxplot)")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

Age vs Fare Scatter Plot (Color-coded by Survived)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Age vs Fare (Scatter Plot)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
